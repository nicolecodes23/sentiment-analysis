## 3.1 Friends TV show google review web scraping

The following code below: <br>
1. Checks if an interim file exists
2. Else starts web scraping from the URL provided
3. Uses Selenium to click and open Google Reviews Button to scrape more reviews
4. Allows users to scroll down the page to load more reviews
5. Extracts the author, rating and review comments
6. Closes the browser

##### Disclaimer
This script relies on Selenium and requires a WebDriver (e.g., ChromeDriver) to interact with web pages. Ensure the appropriate WebDriver is installed and matches your browser version. The instructions to download ChromeDriver will be in the README.ipynb notebook. This project is intended for educational purposes only, and you are responsible for ensuring compliance with the website's Terms of Service.


```python
import os
import pandas as pd

# Define the interim file path
interim_file = "friends_reviews.csv"

# Check if the interim file exists
if os.path.exists(interim_file):
    print(f"Interim file found: {interim_file}")
    df = pd.read_csv(interim_file)
    print(f"Loaded {len(df)} reviews from interim file.")
else:
    print("Interim file not found. Starting live scraping...")

    from selenium import webdriver
    from selenium.webdriver.common.by import By
    import time
    import pandas as pd
    import re
    
    # Set up WebDriver
    driver = webdriver.Chrome()  # Ensure ChromeDriver is in your PATH or specify the path
    
    # Open the specific Google Search page for "Friends reviews"
    driver.get("https://www.google.com/search?q=friends+reviews&rlz=1C5CHFA_enSG1092SG1093&oq=frien&gs_lcrp=EgZjaHJvbWUqDggAEEUYJxg7GIAEGIoFMg4IABBFGCcYOxiABBiKBTISCAEQLhhDGLEDGMkDGIAEGIoFMgYIAhBFGDsyBggDEEUYPDIGCAQQRRhBMgYIBRBFGEEyBggGEEUYPDIGCAcQRRg80gEHODQwajBqN6gCALACAA&sourceid=chrome&ie=UTF-8")
    
    # Click the "Google Reviews" link
    try:
        reviews_button = driver.find_element(By.CLASS_NAME, "S8ee5")
        reviews_button.click()
        time.sleep(3)  # Allow time for the modal or new content to load
        print("Loaded website friends reviews")
    except Exception as e:
        print("Error navigating to reviews:", e)
        driver.quit()
        exit()
    
    # Identify the main reviews container
    try:
        review_modal = driver.find_element(By.CSS_SELECTOR, 'div[jsname="O6ped"]')  # Main container for all reviews
    except Exception as e:
        print("Error locating the review modal:", e)
        driver.quit()
        exit()
    
    # Scroll to dynamically load more reviews
    previous_count = 0
    scroll_attempts = 0
    max_scroll_attempts = 20  # Maximum number of scrolls if no new reviews are loaded

    while scroll_attempts < max_scroll_attempts:
        # Scroll the modal
        driver.execute_script("arguments[0].scrollTop = arguments[0].scrollHeight", review_modal)
        time.sleep(2)  # Allow time for reviews to load
    
        # Check if new reviews have loaded
        elements = driver.find_elements(By.CSS_SELECTOR, 'div[jsname="HeNW9"]')
        if len(elements) > previous_count:  # New reviews detected
            previous_count = len(elements)
            scroll_attempts = 0  # Reset scroll attempts
            print(f"Loaded reviews: {len(elements)}")
        else:
            scroll_attempts += 1  # Increment attempts if no new reviews load
    
    print("Finished scrolling. Total reviews loaded:", len(elements))
    
    # Extract reviews
    reviews = []
    try:
        # Loop through the dynamically loaded review elements
        for element in elements:
            # Extract rating from aria-label attribute
            try:
                star_rating_container = element.find_element(By.CLASS_NAME, "ePCv9")  # Access star-rating container
                rating_raw = star_rating_container.find_element(By.CSS_SELECTOR, "div[aria-label]").get_attribute("aria-label")
                # Clean and extract numeric rating using regex
                rating_match = re.search(r"\d+(\.\d+)?", rating_raw)
                rating = rating_match.group() if rating_match else None
            except Exception:
                rating = None
    
            # Extract the comment text directly from the span inside the div with the jsname
            try:
                comment_container = element.find_element(By.CSS_SELECTOR, "div[jsname='QUIPvd'] span")
                if "display: none;" in element.find_element(By.CSS_SELECTOR, "div[jsname='QUIPvd']").get_attribute("style"):
                    comment_text = comment_container.get_attribute("innerText")
                else:
                    comment_text = comment_container.text
            except Exception:
                comment_text = None
    
            # Extract author details
            try:
                nested_elements = element.find_elements(By.CSS_SELECTOR, 'div[jsname="mWaNFd"].FmcO7b')
                review_text = nested_elements[0].text if nested_elements else None
            except Exception:
                review_text = None
    
            if review_text:  # Add to reviews only if author details exist
                reviews.append({"authors": review_text, "ratings": rating, "comments": comment_text})
    
    except Exception as e:
        print(f"Error extracting reviews: {e}")
    
    # Save reviews to a CSV file
    if reviews:
        df = pd.DataFrame(reviews)
        df.to_csv("friends_reviews.csv", index=False)
        print("Reviews saved to friends_reviews.csv")
    else:
        print("No reviews were found or extracted from friends google review.")
    
    # Close the browser
    driver.quit()
```

Next, I chose to extract data from a pdf using pdfplumber library because it offer several advantages for working with tabular data embedded in PDFs. After attempting to extract data with other methods such as OCR, I found that pdfplumber is able to preserve data integrity the best.

```python
#pip install pdfplumber pandas
```

The following code below: <br>
1. Checks if an interim file exists
2. Else starts extracting data from the pdf provided
3. It first starts by loading and getting the first page of the pdf, since the data for age demographics is located there
4. Converts the first table in page 1 to a dataframe
5. Cleans the dataframe by only keeping specific rows
6. Saves the dataframe into a csv for future use.

```python
import pdfplumber
import pandas as pd
import os

# Check if the CSV file already exists
friends_age_csv = "friends_age.csv"
if not os.path.exists(friends_age_csv):
    print("CSV file not found. Extracting data from the PDF...")

    # Load the PDF
    friends_file_path = "friends_age.pdf"
    with pdfplumber.open(friends_file_path) as pdf:
        # Get the first page
        first_page = pdf.pages[0]
        
        # Extract tables from the page
        tables = first_page.extract_tables()

    # Convert the extracted table into a DataFrame
    # Assuming the relevant table is the first one in the list (adjust index if needed)
    friends_age_data = tables[0]  # Adjust based on which table you need
    friends_age_df = pd.DataFrame(friends_age_data)

    # Clean the DataFrame (set the first row as column headers and drop any unnecessary rows)
    friends_age_df.columns = friends_age_df.iloc[1]  # Set second row as headers
    friends_age_df = friends_age_df[2:9]  # Keep only rows with actual data

    # Save to CSV for future use
    friends_age_df.to_csv(friends_age_csv, index=False)
    print("Data extracted and saved to CSV.")
else:
    print("CSV file already exists.")
    # Load the existing CSV into a DataFrame
    friends_age_df = pd.read_csv(friends_age_csv)

# Display the DataFrame
friends_age_df
```

# 5. Tokenisation and Lemmatization

Using tokenization and lemmatization as part of the preprocessing pipeline for analyzing Google reviews aims to maximize the accuracy and interpretability of sentiment analysis.<br>

Tokenization, achieved with nltk.tokenize.word_tokenize, breaks the text into individual words or tokens, enabling granular analysis of linguistic elements that contribute to sentiment. By isolating tokens, we can detect patterns, extract meaningful phrases, and compute sentiment at a finer level of detail. Lemmatization, implemented using WordNetLemmatizer, refines this process by reducing words to their base or dictionary forms, ensuring that variations of a word, such as "run," "running," and "ran," are treated as semantically equivalent. This is particularly important for emotional analysis, where maintaining the integrity of meaningful words enhances the ability to detect nuanced emotional undertones.
<br><br>
Additionally, lemmatizing tokens as verbs (pos='v') aligns with the dynamic nature of language in reviews, where verbs often express opinions, actions, and sentiments (e.g., "loved," "hate," "enjoyed"). By applying this process to both Friends and Skibidi Toilet review datasets, we establish a standardized and comparable format for cross-analysis. Creating a separate column for processed comments ensures flexibility, allowing us to use both raw and processed data depending on the context of the analysis. This dual-layered approach validates the robustness of our methodology, ensuring both linguistic richness and analytical precision.

```python
#pip install nltk==3.8.1
```

```python
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import nltk

# Download required NLTK data files
nltk.download('punkt')       # For word_tokenize
nltk.download('wordnet')     # For WordNetLemmatizer
nltk.download('omw-1.4')     # For extended WordNet data

# Initialize the lemmatizer
lemmatizer = WordNetLemmatizer()

# Function to tokenize and lemmatize text
def tokenize_and_lemmatize(text):
    tokens = word_tokenize(text)  # Tokenize the text
    lemmatized_tokens = [lemmatizer.lemmatize(token, pos='v') for token in tokens]  # Lemmatize as verbs
    return ' '.join(lemmatized_tokens)  # Join back into a string

# Apply the function to the 'comments' column for both friends and skibidi toilet google reviews
friends_review_data['processed_comments'] = friends_review_data['comments'].apply(tokenize_and_lemmatize)
skib_review_data['processed_comments'] = skib_review_data['comments'].apply(tokenize_and_lemmatize)

skib_review_data.to_csv('cleaned_skib_review.csv', index=False)
friends_review_data.to_csv('cleaned_friends_reviews.csv', index=False)

skib_review_data.head()
```

# 6. Remove stop words

Removing stop words from the processed reviews is a critical step in refining the text data for sentiment analysis and emotional undertone detection. Stop words, such as "the," "is," "and," or "of," are common words that occur frequently in language but do not carry significant semantic meaning or contribute to understanding the sentiment or context of the reviews. By eliminating these words, I am able to focus on words to provide an emotional impact.

```python
from nltk.corpus import stopwords

# Download stop words
nltk.download('stopwords')

# Load English stop words
stop_words = set(stopwords.words('english'))

# Function to remove stop words
def remove_stop_words(text):
    tokens = text.split()  # Split the processed (tokenized and lemmatized) text
    filtered_tokens = [token for token in tokens if token.lower() not in stop_words]  # Remove stop words
    return ' '.join(filtered_tokens)  # Join the tokens back into a string

# Apply the function to remove stop words from the 'processed_comments' column
friends_review_data['final_comments'] = friends_review_data['processed_comments'].apply(remove_stop_words)
skib_review_data['final_comments'] = skib_review_data['processed_comments'].apply(remove_stop_words)

skib_review_data.to_csv('cleaned_skib_review.csv', index=False)
friends_review_data.to_csv('cleaned_friends_reviews.csv', index=False)

friends_review_data.head()
```

## 7.1 Friends TV Show

VADER is specifically designed for sentiment analysis of social media and user-generated content, making it ideal for processing informal, opinion-rich reviews. Its lexicon-based approach incorporates both polarity and intensity, allowing it to effectively handle the nuanced language of reviews, including emotive expressions, punctuation, and even capitalized text for emphasis.<br>

By applying VADER through the SentimentIntensityAnalyzer, I was able to extract a range of sentiment scores (positive, negative, neutral, and a compound score) for each comment. The compound score, in particular, provides a balanced measure of overall sentiment, which I used to classify reviews into meaningful categories: positive, negative, and neutral.
<br>A positive compound score indicates overall positive sentiment, a negative score reflects negative sentiment, and a neutral or near-zero score represents neutral sentiment. This score was used to classify Friends Google reviews into positive, negative, and neutral categories, allowing for a clear understanding of audience sentiment.<br>

I chose this method over alternatives because its rule-based and lexicon-enhanced approach ensures high accuracy for text that includes slang, abbreviations, or informal expressions often present in user reviews. This targeted approach to sentiment analysis enhances the reliability and interpretability of the insights, ensuring the analysis remains focused on the emotional tone and impact of Friends as perceived by its audience.

```python
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer

# Download the VADER lexicon
nltk.download('vader_lexicon')

# Initialize the Sentiment Analyzer
sia = SentimentIntensityAnalyzer()

def analyze_sentiment(text):
    return sia.polarity_scores(text)  # Get sentiment scores
    
# Apply sentiment analysis to the 'final_comments' column
friends_review_data['sentiment_scores'] = friends_review_data['final_comments'].apply(analyze_sentiment)

# Extract sentiment labels from the scores
friends_review_data['sentiment_label'] = friends_review_data['sentiment_scores'].apply(
    lambda x: 'positive' if x['compound'] > 0 else 'negative' if x['compound'] < 0 else 'neutral'
)

# Count the sentiment labels
print(friends_review_data['sentiment_label'].value_counts())

# Save the updated DataFrame to a CSV file
output_file = 'friends_sentiment.csv'
friends_review_data.to_csv(output_file, index=False)

print(f"Sentiment analysis results saved to {output_file}")

```

To address the challenges of sentiment analysis (6 Steps to Effectively Analyze user and customer sentiment, no date), I incorporated enhancements into my sentiment analysis pipeline to better capture context and nuances, including sarcasm, double negatives, and intensifiers.<br><br>
Standard sentiment analysis tools often struggle with context, such as the loss of meaning conveyed by emojis, which can impact the sentiment score when removed. While removing emojis was necessary for text normalization, I mitigated this by focusing on explicit textual expressions of sentiment.<br><br>
To handle double negatives like "not bad" or "not terrible," which can confuse sentiment algorithms, I introduced custom adjustments to recalibrate the compound score, boosting it for positive phrases and penalizing it for negative ones.<br><br>
Additionally, a sarcasm detection function was implemented to flag sarcastic phrases like "yeah, right" or "oh, great," setting their sentiment score to neutral to avoid misclassification.<br><br>
Intensifiers such as "freaking" or "very" were also addressed by amplifying the sentiment score based on their presence. These enhancements ensured that the analysis captured subtle emotional undertones and context, providing a more reliable sentiment classification of Friends Google reviews.

```python
# Function to detect sarcasm
def detect_sarcasm(text):
    lower_text = text.lower()
    if "yeah, right" in lower_text or "oh, great" in lower_text:
        return True
    return False

# Function to adjust for double negatives
def adjust_for_double_negatives(text, sentiment):
    lower_text = text.lower()
    if "not bad" in lower_text or "not terrible" in lower_text:
        sentiment['compound'] = max(sentiment['compound'], 0.5)  # Boost as positive
    elif "not good" in lower_text or "not great" in lower_text:
        sentiment['compound'] = min(sentiment['compound'], -0.5)  # Penalize as negative
    return sentiment

# Function to adjust for intensifiers
def adjust_for_intensifiers(text, sentiment):
    lower_text = text.lower()
    if "freaking" in lower_text or "very" in lower_text:
        sentiment['compound'] = min(sentiment['compound'] + 0.2, 1.0)  # Amplify positive
    elif "extremely bad" in lower_text:
        sentiment['compound'] = max(sentiment['compound'] - 0.2, -1.0)  # Amplify negative
    return sentiment

# Enhanced sentiment analysis function
def analyze_sentiment_enhanced(text):
    # Detect sarcasm
    if detect_sarcasm(text):
        return {'compound': 0, 'sarcasm': True}  # Neutral compound, marked as sarcasm
    
    # Initial sentiment analysis
    sentiment = sia.polarity_scores(text)
    
    # Adjustments for nuances
    sentiment = adjust_for_double_negatives(text, sentiment)
    sentiment = adjust_for_intensifiers(text, sentiment)
    
    # Add sarcasm flag
    sentiment['sarcasm'] = False
    return sentiment

# Apply enhanced sentiment analysis to the 'final_comments' column
friends_review_data['enhanced_sentiment_scores'] = friends_review_data['final_comments'].apply(analyze_sentiment_enhanced)

# Extract sentiment labels from the enhanced scores
friends_review_data['enhanced_sentiment_label'] = friends_review_data['enhanced_sentiment_scores'].apply(
    lambda x: 'positive' if x['compound'] > 0 else 'negative' if x['compound'] < 0 else 'neutral'
)

# Check for sarcasm
friends_review_data['is_sarcastic'] = friends_review_data['enhanced_sentiment_scores'].apply(lambda x: x.get('sarcasm', False))

# Save the enhanced results to a CSV file
enhanced_output_file = 'friends_sentiment.csv'
friends_review_data.to_csv(enhanced_output_file, index=False)

print(f"Enhanced sentiment analysis results saved to {enhanced_output_file}")
print(friends_review_data['sentiment_label'].value_counts())

```

```python
import pandas as pd
import re
from collections import defaultdict

# Load your dataset
skib_review_data = pd.read_csv('skib_sentiment.csv')

# Define a function to process and analyze emotional themes
def analyze_emotional_themes(dataframe, comments_column, emotional_themes):
    # Initialize counters for each theme
    theme_counts = {theme: 0 for theme in emotional_themes.keys()}
    
    # Process each comment
    for comment in dataframe[comments_column].dropna():  # Skip NaN values
        # Tokenize and clean the comment
        words = re.findall(r'\b\w+\b', comment.lower())
        
        # Count keywords in each category
        for word in words:
            for theme, keywords in emotional_themes.items():
                if word in keywords:
                    theme_counts[theme] += 1
    
    return theme_counts

# Emotional themes for Skibidi
themes_skibidi = {
    'absurdity': ['weird', 'chaos', 'random', 'absurd', 'crazy','ohio'],
    'humour': ['funny', 'laugh', 'hilarious', 'joke', 'humour'],
    'fear': ['nightmares', 'scary', 'fear', 'horror'],
    'life_impact': ['life', 'change', 'happy', 'amazing','edging'],
}

# Analyze Skibidi emotional themes
skibidi_emotional_theme_counts = analyze_emotional_themes(skib_review_data, 'final_comments', themes_skibidi)

# Display the results
print("Skibidi Emotional Themes:")
print(skibidi_emotional_theme_counts)
```

- Outliers: While individual outliers are not explicitly visible in a KDE plot, the presence of a long tail extending into the negative range suggests that some reviews have significantly critical sentiment. The minimum score of -0.81 likely reflects these extreme cases, which stand out against the bulk of positive feedback. These outliers could represent unique interpretations, errors in sentiment analysis, or even sarcasm, which are harder for tools like VADER to accurately capture.

- Spread: The KDE curve's sharp peak and narrow width near the positive end confirm a highly concentrated distribution of sentiment scores. However, the curve’s gradual tapering toward the negative end also reflects some variability, supported by the dataset's standard deviation of 0.26. This variability indicates that while most reviews are positive, there are meaningful differences in how audiences express their sentiment.

Unfortunately, it seems like there is also some potential issues with this dataset: 
- Skewness: The heavy positive skew suggests that the dataset predominantly comprises favorable reviews, which might overshadow critical feedback and lead to biased insights.
- Outliers: The considerably large number of negative outliers could result from extreme reviews, potentially influenced by sarcasm, misinterpretation of context, or errors in sentiment scoring.

## 8.8 Episode Duration of Friends and Skibidi Toilet

After analyzing the emotional undertones of Skibidi Toilet and Friends, I decided to examine their episode durations to explore how length influences the type of escapism each show offers.Simultaneously, I wanted to prove that formats of episodes influences ecapism patterns.<br>

Longer episodes, as seen in Friends, provide more room for storytelling, character development, and emotional engagement. This aligns with the comfort and nostalgia Friends evokes, as the extended runtime allows viewers to immerse themselves in the characters’ lives, offering stability and a feel-good escape. 
<br>
In contrast, the short, fast-paced episodes of Skibidi Toilet cater to instant gratification, delivering quick bursts of absurdity, humor, and chaos. The brevity of each episode reflects modern viewing trends, where audiences seek highly engaging, bite-sized content that can be consumed quickly. 
<br><br>
Analyzing episode durations helps connect the emotional themes of each show with their structural formats, offering deeper insight into how these differences shape the escapist experience for their audiences.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Prepare durations in minutes for plotting
friends_duration_data['Duration (min)'] = friends_duration_data['Duration']
skib_duration_data['Duration (min)'] = skib_duration_data['Duration'].apply(
    lambda x: int(x.split(':')[0]) * 60 + int(x.split(':')[1])
) / 60

# Combine the datasets for a violin plot
combined_data = pd.DataFrame({
    'Duration (min)': pd.concat([friends_duration_data['Duration (min)'], skib_duration_data['Duration (min)']]),
    'Category': ['Friends'] * len(friends_duration_data) + ['Skibidi Toilet'] * len(skib_duration_data)
})

# Calculate averages for annotation
average_friends = friends_duration_data['Duration (min)'].mean()
average_skibidi = skib_duration_data['Duration (min)'].mean()

# Plot the violin plot with `hue` assignment
plt.figure(figsize=(10, 6))
sns.violinplot(x='Category', y='Duration (min)', data=combined_data, hue='Category', palette='inferno', dodge=False, legend=False)

# Annotate average values
plt.text(0, average_friends + 1, f'Avg: {average_friends:.1f} min', ha='center', color='white', fontsize=12)
plt.text(1, average_skibidi + 1, f'Avg: {average_skibidi:.1f} min', ha='center', color='white', fontsize=12)

# Customize the plot
plt.title('Comparison of Episode Duration Distributions', fontsize=16)
plt.xlabel('Category', fontsize=12)
plt.ylabel('Duration (Minutes)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.5)

# Display the plot
plt.tight_layout()
plt.show()
```

As seen from the violin plot above, Friends TV Show has a significantly longer episode duration(23.2 mins), while Skibidi Toilet Series averages to 2.8 min, significantly lesser than Friends.
